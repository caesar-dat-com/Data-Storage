```python
import psycopg2

try:
    # Establecer la conexión
    connection = psycopg2.connect(
        user="tu_usuario",
        password="tu_contraseña",
        host="127.0.0.1",
        port="5432",
        database="nombre_de_tu_base_de_datos"
    )

    cursor = connection.cursor()
    print("Conexión exitosa con la base de datos PostgreSQL")

except (Exception, psycopg2.Error) as error:
    print("Error al conectar con PostgreSQL", error)
```

```python
create_tables_query = '''
CREATE TABLE Usuario (
    ID_usuario SERIAL PRIMARY KEY,
    Nombre VARCHAR(50) NOT NULL,
    Apellido VARCHAR(50) NOT NULL,
    Correo_electronico VARCHAR(100) UNIQUE NOT NULL,
    Nombre_de_usuario VARCHAR(50) UNIQUE NOT NULL,
    Contraseña VARCHAR(100) NOT NULL,
    Ciudad_de_origen VARCHAR(50),
    Departamento_de_origen VARCHAR(50),
    Pregunta_secreta VARCHAR(255),
    Respuesta_secreta VARCHAR(255)
);

CREATE TABLE Contraseña (
    ID_contraseña SERIAL PRIMARY KEY,
    ID_usuario INT REFERENCES Usuario(ID_usuario),
    Contraseña VARCHAR(100) NOT NULL,
    Fecha_de_creacion TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
    Estado BOOLEAN DEFAULT TRUE
);

CREATE TABLE Documento (
    ID_documento SERIAL PRIMARY KEY,
    Nombre_del_documento VARCHAR(255) NOT NULL,
    Descripcion TEXT,
    Fecha_de_publicacion TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
    URL TEXT NOT NULL,
    Visibilidad BOOLEAN DEFAULT TRUE
);

CREATE TABLE Autor (
    ID_autor SERIAL PRIMARY KEY,
    Nombre VARCHAR(50) NOT NULL,
    Apellido VARCHAR(50) NOT NULL
);

CREATE TABLE Categoria (
    ID_categoria SERIAL PRIMARY KEY,
    Nombre_de_categoria VARCHAR(100) NOT NULL
);

CREATE TABLE Valoracion (
    ID_valoracion SERIAL PRIMARY KEY,
    ID_usuario INT REFERENCES Usuario(ID_usuario),
    ID_documento INT REFERENCES Documento(ID_documento),
    Valor INT CHECK (Valor >= 1 AND Valor <= 5),
    Fecha_de_valoracion TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);

CREATE TABLE Comentario (
    ID_comentario SERIAL PRIMARY KEY,
    ID_usuario INT REFERENCES Usuario(ID_usuario),
    ID_documento INT REFERENCES Documento(ID_documento),
    Texto_del_comentario TEXT NOT NULL,
    Fecha_de_publicacion TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);

CREATE TABLE Respuesta (
    ID_respuesta SERIAL PRIMARY KEY,
    ID_comentario INT REFERENCES Comentario(ID_comentario),
    ID_usuario INT REFERENCES Usuario(ID_usuario),
    Texto_de_la_respuesta TEXT NOT NULL,
    Fecha_de_publicacion TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);

CREATE TABLE Historial (
    ID_historial SERIAL PRIMARY KEY,
    ID_usuario INT REFERENCES Usuario(ID_usuario),
    ID_documento INT REFERENCES Documento(ID_documento),
    Fecha_de_visualizacion TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);

CREATE INDEX idx_usuario_historial ON Historial(ID_usuario);
'''

try:
    # Ejecutar el script SQL
    cursor.execute(create_tables_query)
    connection.commit()
    print("Tablas y relaciones creadas con éxito en PostgreSQL")

except (Exception, psycopg2.Error) as error:
    print("Error al ejecutar el script SQL", error)

finally:
    # Cerrar la conexión
    if connection:
        cursor.close()
        connection.close()
        print("Conexión con PostgreSQL cerrada")
```

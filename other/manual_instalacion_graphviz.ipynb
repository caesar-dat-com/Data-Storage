# 📚 Manual para la Instalación de Graphviz en Diferentes Sistemas Operativos
Este manual te guiará en el proceso de instalación de Graphviz en tu sistema operativo preferido. Además, aprenderás cómo instalar el módulo de Python para trabajar con Graphviz desde código.

## 1. Instalación de Graphviz

### 1.1 En Ubuntu/Debian
Si estás usando Ubuntu o una distribución basada en Debian, ejecuta el siguiente comando en tu terminal:
```bash
sudo apt-get install graphviz
```

### 1.2 En Fedora
Si utilizas Fedora, instala Graphviz con el siguiente comando:
```bash
sudo dnf install graphviz
```

### 1.3 En macOS
Para macOS, si tienes Homebrew instalado, puedes instalar Graphviz usando este comando:
```bash
brew install graphviz
```

### 1.4 En Windows
En Windows, puedes instalar Graphviz utilizando Chocolatey. Primero, asegúrate de tener [Chocolatey](https://chocolatey.org/install) instalado y luego ejecuta:
```bash
choco install graphviz
```

## 2. Instalación del Módulo Graphviz en Python
Para poder trabajar con Graphviz desde Python, es necesario instalar el módulo `graphviz` mediante pip. Abre tu terminal y ejecuta el siguiente comando:
```bash
pip install graphviz
```

## 3. Verificación de la Instalación
Una vez que hayas instalado Graphviz y el módulo de Python, puedes verificar la instalación ejecutando el siguiente código en un entorno Python:
```python
import graphviz
dot = graphviz.Digraph(comment='Test')
dot.node('A', 'Start')
dot.node('B', 'End')
dot.edge('A', 'B', 'Edge')
dot.render('test-output/round-table.gv', view=True)
```
Este código creará un archivo `.gv` que puedes visualizar para confirmar que Graphviz está funcionando correctamente.

¡Listo! Ahora deberías tener Graphviz instalado y funcionando en tu sistema. 😊
